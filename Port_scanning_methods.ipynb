TCP Connect Scan

```python
import socket

def tcp_connect_scan(target, port):
    try:
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.settimeout(1)

        result = s.connect_ex((target, port))

        if result == 0:
            print(f"[+] Port {port} is OPEN")
        else:
            print(f"[-] Port {port} is CLOSED")

        s.close()
    except Exception as e:
        print(f"Error: {e}")

# Example usage
target = "127.0.0.1"
for port in range(10, 30):
    tcp_connect_scan(target, port)
```

```python
from scapy.all import IP, TCP, sr1, send

def syn_scan(target, port):
    packet = IP(dst=target)/TCP(dport=port, flags="S")

    response = sr1(packet, timeout=1, verbose=0)

    if response is None:
        print(f"[?] Port {port} is FILTERED")

    elif response.haslayer(TCP):
        if response[TCP].flags == 18:  # SYN-ACK
            print(f"[+] Port {port} is OPEN")

            # Send RST to close connection
            rst = IP(dst=target)/TCP(dport=port, flags="R")
            send(rst, verbose=0)

        elif response[TCP].flags == 20:  # RST-ACK
            print(f"[-] Port {port} is CLOSED")

# Example
target = "127.0.0.1"
for port in range(10, 30):
    syn_scan(target, port)
```

UDP Scan

```python
from scapy.all import IP, UDP, sr1, ICMP

def udp_scan(target, port):
    packet = IP(dst=target)/UDP(dport=port)

    response = sr1(packet, timeout=2, verbose=0)

    if response is None:
        print(f"[?] Port {port} is OPEN or FILTERED")

    elif response.haslayer(ICMP):
        if int(response.getlayer(ICMP).type) == 3:
            print(f"[-] Port {port} is CLOSED")

# Example
target = "127.0.0.1"
for port in range(50, 60):
    udp_scan(target, port)
```
